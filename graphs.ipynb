```python
import pandas as pd
import numpy as np
import json
import os
from scipy.signal import medfilt
```

```python
ride_name = 'witches_wheel'
data_folder = 'data/2013-09-28-cedar-point'
ride_folder = os.path.join(data_folder, ride_name)
```

```python
data = pd.read_csv(os.path.join(ride_folder, 'accelerometer_log.txt').format(ride_name), sep=",", header=1)
```

```python
start=2400
end = len(data.time) - 200
plot(data.time[start:end], medfilt(np.sqrt(np.power(data.x[start:end], 2) + np.power(data.y[start:end], 2) + np.power(data.z[start:end], 2)),1))
figure()
plot(data.time[start:end], data.x[start:end], 'r-', data.time[start:end], data.y[start:end], 'g-', data.time[start:end], data.z[start:end], 'b-')
```

```python
notes_fn = os.path.join(ride_folder, 'notes.json')
try:
    notes = json.load(open(notes_fn, 'rb'))
except IOError:
    notes = {}
notes['ride_start'] = start
notes['ride_end'] = end
json.dump(notes, open(notes_fn, 'wb'))
```
